#6. Escala los datos, si consideras necesario

La estandarizacin de datos, en ocasiones tambin conocida como normalizacin, es el proceso de ajustar o adaptar ciertas caractersticas para que los datos se asemejen a un tipo, modelo o normal comn con el objetivo de que su tratamiento, acceso y uso sea ms sencillo para los usuarios o personas que dispongan de ellos.

El propio concepto de datos abiertos implica una bsqueda de la estandarizacin en el uso de la informacin en formato abierto para que pueda utilizarse y ser reutilizada por parte de la ciudadana.

Por tanto, podra decirse que la estandarizacin es la forma en la que todas las personas pueden comparar y consultar datos y encontrar siempre la informacin que necesitan, teniendo certeza de uniformidad en la forma en la que lo encontrarn.

Normalizacin de puntuacin Z (Estandarizacin) - convierte los datos del conjunto a tener una media de 0 y una desviacion estandar de 1 

 $$ x_i^{new} = \cfrac{x_i^{old} - \mu}{\sigma}$$ 

 La frmula se define como  la resta entre el valor x en turno de nuestro arreglo menos el promedio de los valores del arreglo x, todo entre la desviacin etandar del arreglo x. ((X-xm)/sd)

 Cabe resaltar que la estandarizacin es de los metodos ms utilizados para hacer PCAs.

```python
#Extraer Variables Numerica
numericas = ['X1', 'X5', 'X12', 'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X19', 'X20', 'X21', 'X22', 'X23']
df_main_num = df_main[numericas]
df_main_num
```

```python
#Definimos una funcin de estandarizacin o normalizacin.

def standarization(X):
  xm = np.mean(X)
#El metodo statisctics.pstdev(), calcula la desviacin estandar de un arreglo
  sd =sts.pstdev(X)
  return ((X-xm)/sd)
```

```python
#A continuacin normalizamos  todas las columnas en el rango de 0-1, usando la funcin previamente definida normalize
#esta funcin re escala la informacin de valores entre 0 y 1
#Nuestro nuevo dataframe de valores normalizados se llamara df_2 normalizew
columnas = df_main_num.columns
print(columnas)
df_main_estandarizado = df_main_num.copy()
for n in columnas:
  df_main_estandarizado[n] = standarization(df_main_num[n])
df_main_estandarizado
```

#7. Reduce las dimensiones con PCA, si consideras necesario.
1. Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploracin de los datos la varianza > 70%
1. Indica la importancia de las variables en cada componente

```python
#Generamos un nuevo dataset llamado varianza, el cual contendra la covarianza cov del data set previamente normalizado
#Considerando que el dataset se traspone con el .T, por lo que toma una forma mas simple de analisis en cuanto cantidad de columnas y filas

matriz_cov = df_main_estandarizado.corr()
```

```python
#Generamos un Heatmap para la matriz de correlacion llamada varianza
sns.set(rc={'figure.figsize':(20,20)}) 
sns.heatmap(matriz_cov, cmap = 'YlGnBu', annot=True,)
```

```python
#Procedmeos a hacer un estudio PCA de 5 componentes minimos a ver si cumplimos con el 70% de la varianza total
componente_minimos = 5
pca = PCA(n_components=componente_minimos)
pca_data = pca.fit_transform(df_main_estandarizado)


print('Varainza de cada componente minimo\n')
pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pca.explained_variance_),
                             'Proportion of variance': pca.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pca.explained_variance_ratio_)
                             })
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df
#Sacamos valores estadisticos de los pesos.
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4)

```

Podemos ver que cumplimos con la varianza pedida del 70%

Ahora podemos notar que con 5 componentes mnimos pudimos llegar al objetivo de tener un mnimo de 70% de la varianza total.

```python
#sacamos varianzas de acuerdo a los pesos
pcsComponents_df = pd.DataFrame(pca.components_.transpose(), 
                                columns = pcsSummary_df.columns,
                                index = df_main_estandarizado.columns
                                )
pcsComponents_df
```


Tomando en cuenta que estamos usando el dataset limpio y que se redujo todo a solo 5 columnas de datos, valores que representan mayores pesos en estas varianzas de acuerdo a la tabla son principales pesos serian las siguientes:

```python
names = pcsComponents_df.columns
x = 1
for n in names:
 print(f"Variable con mas peso en PC{x}:", pcsComponents_df[n].idxmax(axis=0, skipna=False))
 x+=1
```

```python
PC_components = pcsSummary_df.columns
#PC_components

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pca.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components, 
                 y = np.cumsum(pca.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Histograma')
plt.xlabel('N-th Componentes principales')
plt.ylabel('Varianza explicada')
plt.ylim(0, 1)
plt.show()
```
